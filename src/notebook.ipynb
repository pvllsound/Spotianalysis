```python
import requests
import os
from dotenv import load_dotenv
load_dotenv()
CLIENT_ID = os.getenv('CLIENT_ID')
CLIENT_SECRET = os.getenv('CLIENT_SECRET')
AUTH_URL = "https://accounts.spotify.com/api/token"
```

```python
auth_response = requests.post(AUTH_URL,{'grant_type': 'client_credentials','client_id': CLIENT_ID, 'client_secret': CLIENT_SECRET})
```

```python
auth_response_data = auth_response.json()
```

```python
access_token = auth_response_data['access_token']
```

```python
auth_response_data
```

```python
# headers = {'Authorization': 'Bearer {token}'.format(token=access_token)}
headers = {
    'Authorization': 'Bearer {token}'.format(token=access_token)
}

# base URL of all Spotify API endpoints
BASE_URL = 'https://api.spotify.com/v1/'

# Track ID from the URI
track_id = '6y0igZArWVi6Iz0rj35c1Y'

# actual GET request with proper header
r = requests.get(BASE_URL + 'audio-features/' + track_id, headers=headers)
```

```python
r = r.json()
r
```

```python
artist_id = '36QJpDe2go2KgaRleHCDTp'

result = requests.get(BASE_URL + 'artists/' + artist_id + '/albums', headers=headers, params={'include_groups': 'album', 'limit': 50})

data = result.json()

data
```

```python
for album in data['items']:
    print(album['name'], ' ', album['release_date'])
```
